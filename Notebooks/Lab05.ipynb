```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.head()
```

```python
#Mostrar dimensiones
df.shape
```

```python
#Nombres de columnas
df.columns
```

```python
#Tipos de datos
df.dtypes
```

```python
#LIMPIEZA
#Valores nulos
df.isnull().sum()
```

```python
#Porcentaje
(df.isnull().sum()/len(df))*100
```

```python
#Registros duplicados
df.duplicated().sum()
```

```python
#Eliminar duplicados
df = df.drop_duplicates()
```

```python
#valores faltantes
df['director'] = df['director'].fillna('Sin director')
df['cast'] = df['cast'].fillna('Sin elenco')
df['country'] = df['country'].fillna('Desconocido')
```

```python
#Generar Graficas
sns.countplot(x='type', data=df)
plt.title("Películas vs Series")
plt.show()
```

```python
#Contenido por año
df['release_year'].value_counts().head(10).plot(kind='bar')
plt.title("Años con más contenido")
plt.show()
```

```python
#Países con más contenido
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Países con más contenido")
plt.show()
```

```python
#Clasificaciones más comunes
sns.countplot(y='rating', data=df,
order=df['rating'].value_counts().index[:10])
plt.title("Clasificaciones más comunes")
plt.show()
```

```python
#CONCLUSIONES
#. Netflix tiene más películas que series en su catálogo.

#. Estados Unidos es el país con mayor cantidad de contenido.

#. La mayoría de los registros pertenecen a clasificaciones para adolescentes y adultos.

#. El análisis permitió identificar valores faltantes y mejorar la calidad de los datos mediante limpieza y transformación.
```
